## Adding Diff Shapes to the  Image

```python
img_ones = np.ones((512,512,3))
```

```python
img_ones
```

```python
img_ones.shape
```

```python
cv2.imshow('ones image',img_ones)
cv2.waitKey(-1)
# we get white image here
```

```python
img_rect = cv2.rectangle(img_ones, pt1=(50,50), pt2=(300,300), color=(255,0,0),thickness=3)

# we can also draw circle,line etc.
```

```python
cv2.imshow('rect image',img_rect)
cv2.waitKey(-1)
```

## Adding Text to Image

```python
cv2.imshow('ones image',img_ones)
cv2.waitKey(-1)
```

```python
img_text = cv2.putText(img_ones, text = 'Hello All', org = (50,50), fontFace= cv2.FONT_ITALIC, color = (0,255,0), fontScale=2)
```

```python
cv2.imshow('text image',img_text)
cv2.waitKey(-1)
```

## Infinite Image Frame

```python
import cv2
import numpy as np
img_infinite = np.ones((512,512,3))
```

```python
while True:
    cv2.imshow('image window', img_infinite)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

## Image Events

```python
import cv2
import numpy as np
img_events = np.zeros((512,512,3))
```

```python
def catchevent(event, x, y, flag, param):
    
    if event == 0:
        print('Mouse is Moving')
        
    elif event == 1:
        print('Clicked Down')
        
    else:
        print('Released Mouse')
   

    

cv2.namedWindow(winname='zero_window')
cv2.setMouseCallback('zero_window', catchevent)



while True:
    cv2.imshow('zero_window', img_events)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```
